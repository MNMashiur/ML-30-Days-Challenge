# Missing Value

```python
import pandas as pd
import numpy as np
```

```python
data= {
    'Name':['John','Anna','Mike','Sara','Tom'],
    'Age':[28,np.nan,35,29,np.nan],
    'Salary':[5000,6000,np.nan,7000,8000],
}

df=pd.DataFrame(data)

df
```

```python
df.isnull().sum()
```

```python
#drop

df_drop=df.dropna()

df_drop
```

```python
df_stat= df.copy()

df_stat['Age'].fillna(df_stat['Age'].mean())

df_stat
```

```python
df_stat['Salary'].fillna(df_stat['Salary'].median())

df_stat
```

# Duplicate Handling

```python
data= {
    'Name':['John','Anna','Mike','Sara','John'],
    'Age':[28,22,35,29,28],
    'City': ['Dhaka', 'Chittagong', 'Dhaka', 'Khulna', 'Dhaka']
}

df=pd.DataFrame(data)

df
```

```python
#check duplicate
df.duplicated()
```

```python
#rmv duplicate
df_dup= df.drop_duplicates()

df_dup
```

# Pivot table

```python
data= {
    'Name': ['Riya', 'Alex', 'Toma', 'Karim', 'Nila'],
    'Department': ['CSE', 'CSE', 'EEE', 'EEE', 'CSE'],
    'Gender': ['F', 'M', 'F', 'M', 'F'],
    'Marks': [85,75,80,70,90]
}

df=pd.DataFrame(data)

df
```

```python
pivot=df.pivot_table(values='Marks',
                     index='Department',
                     columns='Gender',
                     aggfunc='mean')

pivot
```
