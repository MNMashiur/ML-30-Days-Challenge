**Line Chart**

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,3,10,12]

plt.plot(x,y, color="red", marker="o", linestyle="--")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Graph")
plt.show()
```

**Multiple line**

```python
x=[1,2,3,4,5,6]
y=[2,4,6,8,10,12]
y2=[1,4,9,16,25,36]

plt.plot(x,y, label="y=2x")
plt.plot(x,y2, label="y=x^2")
plt.legend()

plt.show()
```

**Bargraph**

```python
player=['Sakib','Tamim', 'Nasir','Riyad','Mushi']
runs=[12000,12500,6000,11000,11500]

plt.bar(player, runs, color='Green')
plt.xlabel('Player')
plt.ylabel('Runs')
plt.title('Player & Their Runs')

plt.show()
```

**Histogram**

```python
runs=[6,8,4,5,8,6,4,7,8,9,6,8,4,5,8,6,4,7,8,9]

plt.hist(runs, bins=4, color='purple', edgecolor='white')
plt.xlabel('Over')
plt.ylabel('Runs')
plt.title('Over & Runs')

plt.show()
```

**PieChart**

```python
size=[20,30,25,25]
labels=['A','B','C','D']

plt.pie(size, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Size & Labels')


plt.show()
```

**BoxPlot**

```python
data=[7,8,6,9,10,12,14,8,7,9,30]

plt.boxplot(data)
plt.title('Outlier Detection')

plt.show()
```

**Multiple FIgure Draw In a Row**

```python
x=[1,2,3,4,5,6]
y=[2,4,6,8,10,12]
y2=[1,4,9,16,25,36]

#plot 1
plt.subplot(1,2,1)
plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph 1")
plt.show

#plot 2
plt.subplot(1,2,2)
plt.plot(x, y2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph 2")


plt.show()
```
