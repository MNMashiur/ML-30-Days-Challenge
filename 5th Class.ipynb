Argument and parameter

```python
def greet(name):
  print('Hello '+name)

greet('Abdulla')
greet('Mashiur')
```

```python
def about(name, age):
  print(f"I am {name} and I am {age} years old")

about("Mashiur", 24)
```

Arbitary Argument

```python
def info(*arg):   #   * for unknown number of argument
  for i in arg:
    print(i)

info("Mashiur", 28, "EWU", 3.68)
```

keyword Argument

```python
def about(age, name, GPA):
  print(name, age, GPA)

about(name= "Mashiur", GPA= 5.00, age= 24)   #accessing the funciton by keyword
```

Arbitary Keyword

```python
def about(**arg):
  for i, j in arg.items():
    print(i, ":", j)

about(name= "Mashiur", age= 24, institute= "EWU")
```

Default Parameter

```python
def country(name= "BD"):
  print(name)

country("Japan")
country()
```

Return Function

```python
def sum(a, b):
  return a+b

data= sum(3, 5)
print(data)
```
