**Combining two Dataset**

```python
df1= pd.DataFrame({'Fruits' : ['Mango', 'Banana', 'Orange', 'Apple','grape'],
      'Quantity': [60,40,75,100,65]})

df1
```

```python
df2= pd.DataFrame({'Fruits' : ['Apple', 'Orange', 'Papaya', 'Pinaple','Mango'],
      'Price': [120,60,30,70,50]})

df2
```

```python
df3= pd.merge(df1,df2, how= 'left', on= 'Fruits')

df3
```

```python
df4= pd.merge(df1,df2, how= 'inner', on= 'Fruits')

df4
```

```python
df5= pd.merge(df1,df2, how= 'outer', on= 'Fruits')

df5
```

```python
df6= pd.merge(df1,df2, how= 'right', on= 'Fruits')

df6
```

**Concat**

```python
df3=pd.concat([df1,df2], axis=0, ignore_index=True)

df3
```
